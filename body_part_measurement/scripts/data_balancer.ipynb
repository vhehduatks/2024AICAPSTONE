```python
import os
import glob
import json
import pandas as pd
import random
```

```python
path_base = "D:\\data\\body_parts_measurement\\train"
list_path_csvs = glob.glob(path_base + "\\*\\csv\\M*.csv")
len(list_path_csvs)
```

```python
dict_ages = dict()
```

```python
for path_csv in list_path_csvs :
    username = os.path.basename(path_csv).split(".")[0]
    df = pd.read_csv(path_csv, encoding = "ISO-8859-1")
    body_parts_measurement = list(df.iloc[[1]].values[0][3:-1])
    age = df.iloc[[1]].values[0][-2]
    dict_ages[username] = int(float(age))
```

```python
dict_ages
```

```python
dict_ranges = dict()

for key, value in dict_ages.items() :
    key_range = int(value / 10)
    if key_range not in dict_ranges :
        dict_ranges[key_range] = []
        dict_ranges[key_range].append(key)
    else :
        dict_ranges[key_range].append(key)
```

```python
list_train = []
list_validation = []
list_test = []

ratio_train = 8
ratio_validation = 1
ratio_test = 1

for key, value in dict_ranges.items() :
    random.shuffle(value)
    
    idx_train = int(len(value)/10*(ratio_train))
    idx_train_and_validation = int(len(value)/10*(ratio_train+ratio_validation))
    list_train.extend(value[:idx_train])
    list_validation.extend(value[idx_train:idx_train_and_validation])
    list_test.extend(value[idx_train_and_validation:])

list_train.sort()
list_validation.sort()
list_test.sort()
print(len(list_train))
print(len(list_validation))
print(len(list_test))
```
