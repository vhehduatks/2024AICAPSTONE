```python
import cv2
from ultralytics import YOLO

# YOLO 모델 로드
model = YOLO('yolo11n-pose.pt')

# 웹캠 초기화
cap = cv2.VideoCapture(0)

if not cap.isOpened():
	print("웹캠을 열 수 없습니다.")
	exit()

while True:
	# 프레임 읽기
	ret, frame = cap.read()
	if not ret:
		print("프레임을 읽을 수 없습니다.")
		break
	
	# 모델 추론
	results = model.predict(frame, conf=0.5)

	annotated_frame = results[0].plot()

	for result in results:
		if hasattr(result, 'keypoints') and result.keypoints is not None:
			keypoints_data = result.keypoints.data[0]  # 첫 번째 객체의 키포인트 데이터

			try:
				# 신뢰도 기준으로 부위 확인
				head_conf = keypoints_data[0, 2]  # 머리 신뢰도
				left_hand_conf = keypoints_data[9, 2]  # 왼손 신뢰도
				right_hand_conf = keypoints_data[10, 2]  # 오른손 신뢰도
				left_leg_conf = keypoints_data[15, 2]  # 왼쪽 다리 신뢰도
				right_leg_conf = keypoints_data[16, 2]  # 오른쪽 다리 신뢰도

				# 모든 부위가 신뢰도 0.5 이상일 경우
				if (
					head_conf > 0.5 and
					left_hand_conf > 0.5 and
					right_hand_conf > 0.5
					# left_leg_conf > 0.5 and
					# right_leg_conf > 0.5
				):
					# 바운딩 박스 정보 가져오기
					for box in result.boxes.data.tolist():
						x1, y1, x2, y2, confidence, class_id = map(int, box[:6])
						if class_id == 0:  # 사람 클래스
							# 이미지 크롭
							cropped_img = frame[y1:y2, x1:x2]

							# 크롭된 이미지 저장
							file_name = f"cropped_person_{x1}_{y1}.jpg"
							cv2.imwrite(file_name, cropped_img)
							print(f"이미지 저장: {file_name}")
			except:
				pass	

	# 추론 결과 프레임 출력
	cv2.imshow("YOLOv8 Inference", annotated_frame)

	# 'q' 키를 누르면 종료
	if cv2.waitKey(1) & 0xFF == ord('q'):
		break

# 자원 해제
cap.release()
cv2.destroyAllWindows()
```
